```python
df_consumption = pd.read_csv('energy_cons.csv')
# df_consumption
df_consumption = df_consumption.set_index("Country")
year = [ 
         '2000', '2001', '2002', '2003', '2004', '2005',
        '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014',
        '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
# year = [item + "年" for item in a]
df_consumption.columns = year
# df_consumption.info()
```

```python
data = []
for column in df_consumption.columns:
    dict_temp = {}
    dict_temp["time"] = column
    dict_temp["data"] = []
    for index, value in df_consumption[column].items():

        dict_temp2 = {}
        dict_temp2["name"] = index
        dict_temp2["value"] = [value,value/(df_consumption[column].sum()),index]
        dict_temp["data"].append(dict_temp2)
    data.append(dict_temp)
```

```python
total_num = [ df_consumption[column].sum() for column in df_consumption.columns]
# print(total_num)
```

```python
time_list = [str(d) for d in range(2000, 2023)]
# timelist 

# total_num = [ df_consumption[column].sum() for column in df_consumption.columns]
maxNum = 44276


minNum = 0
# data = data2
# total_num = [7569, 7635, 7650, 7753, 7791, 8001, 8191, 8257, 8281, 8470, 8679, 8743, 8924, 9255, 9662, 9939, 10218, 10493, 10598, 10503, 11096, 11294, 11432, 11607, 11738, 11735, 11812, 12046, 12362, 12502, 12133, 12746, 12966]
# time_list = time_list2
total_num = [ round(df_consumption[column].sum(),2) for column in df_consumption.columns]
print(data)

def get_year_chart(year: str):
    map_data = [
        [[x["name"], x["value"]] for x in d["data"]] for d in data if d["time"] == year
    ][0]
    min_data, max_data = (minNum, maxNum)
    data_mark: List = []
    i = 0
    for x in time_list:
        if x == year:
            data_mark.append(total_num[i])
        else:
            data_mark.append("")
        i = i + 1

    map_chart = (
        Map()
        .add(
            maptype = 'world',
            series_name="",
            data_pair=map_data,
            zoom=1,
            center=[119.5, 34.5],
            is_map_symbol_show=False,
            itemstyle_opts={
                "normal": {"areaColor": "#323c48", "borderColor": "#404a59"},
                "emphasis": {
                    "label": {"show": Timeline},
                    "areaColor": "rgba(255,255,255, 0.5)",
                },
            },
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title="" + str(year) + " Total Energy Consumption",
                subtitle="",
                pos_left="center",
                pos_top="top",
                title_textstyle_opts=opts.TextStyleOpts(
                    font_size=25, color="rgba(255,255,255, 0.9)"
                ),
            ),
            tooltip_opts=opts.TooltipOpts(
                is_show=True,
                formatter=JsCode(
                    """function(params) {
                    if ('value' in params.data) {
                        return params.data.value[2] + ': ' + params.data.value[0];
                    }
                }"""
                ),
            ),
            
            # visualmap_opts=opts.VisualMapOpts(
                # is_calculable=True,
                # dimension=0,
                # pos_left="30",
                # pos_top="center",
                # range_text=["High", "Low"],
                # range_color=["lightskyblue", "yellow", "orangered"],
                # textstyle_opts=opts.TextStyleOpts(color="#ddd"),
                # min_=min_data,
                # max_=max_data,
            # ),
        )
        .set_series_opts(
            labelopts = opts.LabelOpts(is_show=False)
        )
    )

    line_chart = (
        Line()
        .add_xaxis(time_list)
        .add_yaxis("", total_num)
        .add_yaxis(
            "",
            data_mark,
            markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")]),
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title="Total Energy Consumption in 2000-2022(TWh)", pos_left="64%", pos_top="5%"
            )
        )
    )
    bar_x_data = [x[0] for x in map_data]
    bar_y_data = [{"name": x[0], "value": x[1][0]} for x in map_data]
    bar = (
        Bar()
        .add_xaxis(xaxis_data=bar_x_data)
        .add_yaxis(
            series_name="",
            y_axis=bar_y_data,
            label_opts=opts.LabelOpts(
                is_show=False, position="right", formatter="{b} : {c}"
            ),
        )
        .reversal_axis()
        .set_global_opts(
            xaxis_opts=opts.AxisOpts(
                max_=maxNum, axislabel_opts=opts.LabelOpts(is_show=False)
            ),
            yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(is_show=False)),
            tooltip_opts=opts.TooltipOpts(is_show=True,formatter="Energy Consumption：<br>{b} : {c}"),
            visualmap_opts=opts.VisualMapOpts(
                is_calculable=True,
                dimension=0,
                pos_left="10",
                pos_top="60%",
                range_text=["High", "Low"],
                range_color=["lightskyblue", "yellow", "orangered"],
                textstyle_opts=opts.TextStyleOpts(color="#ddd"),
                min_=min_data,
                max_=max_data,
            ),
        )
    )

    pie_data = [[x[0], x[1][0]] for x in map_data]
    pie = (
        Pie()
        .add(
            series_name="",
            data_pair=pie_data,
            radius=["15%", "35%"],
            center=["80%", "82%"],
            itemstyle_opts=opts.ItemStyleOpts(
                border_width=1, border_color="rgba(0,0,0,0.3)"
            ),
        )
        .set_global_opts(
            tooltip_opts=opts.TooltipOpts(is_show=True, formatter="{b} {d}%"),
            legend_opts=opts.LegendOpts(is_show=False),
        )
        .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False)
        )
    )

    grid_chart = (
        Grid()
        .add(
            bar,
            grid_opts=opts.GridOpts(
                pos_left="65%", pos_right="80", pos_top="10%", pos_bottom="50%", width="1px", height="1px"
            ),
        )
        .add(
            line_chart,
            grid_opts=opts.GridOpts(
                pos_left="65%", pos_right="80", pos_top="10%", pos_bottom="50%"
            ),
        )
        .add(pie, grid_opts=opts.GridOpts(pos_left="45%", pos_top="60%"))
        .add(
            map_chart, grid_opts=opts.GridOpts(pos_left="200"
            )
            )
    )

    return grid_chart


if __name__ == "__main__":
    timeline = Timeline(
        init_opts=opts.InitOpts(width="100vw", height="100vh", theme=ThemeType.DARK)
    )
    for y in time_list:
        g = get_year_chart(year=y)
        timeline.add(g, time_point=str(y))

    timeline.add_schema(
        orient="vertical",
        is_auto_play=True,
        is_inverse=True,
        play_interval=5000,
        pos_left="null",
        pos_right="5",
        pos_top="20",
        pos_bottom="20",
        width="60",
        label_opts=opts.LabelOpts(is_show=True, color="#fff"),
    )

    timeline.render("地图1.html")
```

# 图7 各国人均能源消耗堆栈图

```python
dta1 = pd.read_csv('per-capita-energy-stacked.csv')
dta1 = dta1.fillna(0)
dta1['Total'] = dta1[['Coal', 'Oil', 'Gas', 'Nuclear', 'Hydro', 'Wind', 'Solar', 'Other renewables']].sum(axis = 1)

dta1['Coal_per'] = (dta1['Coal'] / dta1['Total'] * 100)
dta1['Oil_per'] = (dta1['Oil'] / dta1['Total'] * 100)
dta1['Gas_per'] = (dta1['Gas'] / dta1['Total'] * 100)
dta1['Nuclear_per'] = (dta1['Nuclear'] / dta1['Total'] * 100)
dta1['Hydro_per'] = (dta1['Hydro'] / dta1['Total'] * 100)
dta1['Wind_per'] = (dta1['Wind'] / dta1['Total'] * 100)
dta1['Solar_per'] = (dta1['Solar'] / dta1['Total'] * 100)
dta1['Other_per'] = (dta1['Other renewables'] / dta1['Total'] * 100)
dta1
```

```python
years = list(range(2000, 2023))
countries = ['United\nStates', 'Sweden', 'Germany', 'Japan', 'France', 'China', 'United\nKingdom', 'South\nAfrica', 'Brazil', 'India', 'World']
categories_1 = ['Coal', 'Oil', 'Gas', 'Nuclear', 'Hydro', 'Wind', 'Solar', 'Other renewables']
categories_2 = ['Coal_per', 'Oil_per', 'Gas_per', 'Nuclear_per', 'Hydro_per', 'Wind_per', 'Solar_per', 'Other_per']

colors = ['#404A3C', '#4F4E2E', '#F6EDD9', '#00679D', '#72B2DF', '#72B2DF', '#DC9F53', '#A3AF9E']
#colors = ['#5BB946', '#00A668', '#00907D', '#007881', '#007FB5', '#006AB4', '#006073', '#2F4858']
```

```python
timeline = Timeline()
for year in years:
    bar = Bar()
    bar.add_xaxis(list(countries))
    for i, category in enumerate(categories_1):
        data_list = []
        for country in countries:
            value = dta1[(dta1['Year'] == year) & (dta1['Entity'] == country)][category].values
            data_list.append(value[0] if len(value) > 0 else 0)
        bar.add_yaxis(category, data_list, stack = "stack1",
                       itemstyle_opts = opts.ItemStyleOpts(color = colors[i]),
                       label_opts=None
        )
    #bar.reversal_axis()
    bar.set_global_opts(
        title_opts = opts.TitleOpts(title = f"{year} Energy Distribution per Capita", pos_left = "center"),
        legend_opts = opts.LegendOpts(pos_top = "5%"),
        yaxis_opts = opts.AxisOpts(
            type_ = "value",
            name = "Kwh"
        ),
        tooltip_opts = opts.TooltipOpts(
            trigger = "axis",
            axis_pointer_type = "shadow"
        )
    )

    timeline.add(bar, str(year))

timeline.width = "70%"

timeline.render_notebook()
#timeline.render("stacked_bar_timelinechart.html")
```

# 图8 各国人均能源消耗百分比堆栈图

```python
timeline = Timeline()
for year in years:
    bar = Bar()
    bar.add_xaxis(list(countries))
    for i, category in enumerate(categories_2):
        data_list = []
        for country in countries:
            value = dta1[(dta1['Year'] == year) & (dta1['Entity'] == country)][category].values
            data_list.append(value[0] if len(value) > 0 else 0)
        bar.add_yaxis(category.replace('_per', ''), data_list, stack = "stack1",
                       itemstyle_opts = opts.ItemStyleOpts(color = colors[i]),
                       #label_opts = opts.LabelOpts(formatter = lambda x: f'{x:.2f}%', position = "inside", font_size = 10))
                       label_opts=opts.LabelOpts(is_show=True,
                             formatter = JsCode(
                                 """
                                 function(params) {
                                
                                     return (params.value).toFixed(2)
                                 }
                                 """
                            )))
                       #label_opts=opts.LabelOpts(formatter="{c:.2f}%", position="inside", font_size=10))
    
    bar.set_global_opts(
        title_opts = opts.TitleOpts(title = f"{year} Energy Distribution per Capita", pos_left = "center"),
        legend_opts = opts.LegendOpts(pos_top = "5%"),
        yaxis_opts = opts.AxisOpts(
            type_ = "value",
            max_ = 100,
            axislabel_opts = opts.LabelOpts(formatter = "{value}%"),
            name = "Percentage"
        ),
        tooltip_opts = opts.TooltipOpts(
            trigger = "axis",
            axis_pointer_type = "shadow"
        )
    )
    
    timeline.add(bar, str(year))
timeline.width = "70%"

timeline.render_notebook()
#timeline.render("stacked_bar_percent_timelinechart.html")
```
